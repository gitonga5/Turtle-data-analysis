##### 【Problem 1 】 Data acquisition &【Problem 2 】 Data coupling

```python
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris()

# Create a Pandas DataFrame with feature data
df = pd.DataFrame(iris.data, columns=["sepal_length", "sepal_width", "petal_length", "petal_width"])

# Add the target variable (species)
df["species"] = iris.target

# Map the numerical target to the actual species names
#df["species"] = df["species"].map({0: "setosa", 1: "versicolor", 2: "virginica"})

# Display the first few rows
print(df.head())
```

##### 【Problem 6 】 Creating a diagram

```python
#Pie Chart – Number of Samples per Label
import matplotlib.pyplot as plt

# Count samples per species
species_counts = df["species"].value_counts()

# Pie chart
plt.figure(figsize=(6, 6))
plt.pie(species_counts, labels=species_counts.index, autopct="%1.1f%%", colors=["lightblue", "orange", "green"])
plt.title("Proportion of Each Iris Species")
plt.show()
```

```python
#Box Plot – Distribution of Each Feature by Label
import seaborn as sns

# Set figure size
plt.figure(figsize=(8, 6))

# Boxplot for sepal length (you can change to other features)
sns.boxplot(x="species", y="sepal_length", data=df)

# Title and show
plt.title("Box Plot of Sepal Length by Species")
plt.show()
```

```python
#Violin Plot – Feature Distribution by Label
# Set figure size
plt.figure(figsize=(8, 6))

# Violin plot for sepal length
sns.violinplot(x="species", y="sepal_length", data=df)

# Title and show
plt.title("Violin Plot of Sepal Length by Species")
plt.show()
```

##### 【Problem 7 】 Confirmation of relationship between characteristic quantities

```python
#Scatter Plots for Feature Combinations
import seaborn as sns
import matplotlib.pyplot as plt

# Set figure size
plt.figure(figsize=(12, 8))

# List of feature pairs
feature_pairs = [("sepal_length", "sepal_width"), 
                 ("sepal_length", "petal_length"), 
                 ("sepal_length", "petal_width"),
                 ("sepal_width", "petal_length"), 
                 ("sepal_width", "petal_width"), 
                 ("petal_length", "petal_width")]

# Loop through feature pairs and create subplots
for i, (x, y) in enumerate(feature_pairs, 1):
    plt.subplot(2, 3, i)
    sns.scatterplot(x=df[x], y=df[y], hue=df["species"], palette="Set1")
    plt.xlabel(x)
    plt.ylabel(y)

plt.tight_layout()
plt.show()
```

```python
#Pairplot – All Scatter Plots at Once
sns.pairplot(df, hue="species", diag_kind="kde", palette="Set1")
plt.show()
```

```python
#Correlation Matrix (Numerical Values)
# Compute correlation matrix
corr_matrix = df.drop(columns=["species"]).corr()

# Display the matrix
print(corr_matrix)
```

```python
#Heatmap of Correlation Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```
